```python
  def make_menu():
    import csv
    import glob

    m = {}

    for f in glob.glob("data/*.csv"):
        k = f.split(".")[0]
        rows = csv.reader(open(f))

        _m = {}
        for itm in rows:
            _m[itm[0]] = [itm[1], itm[2]]
        m[k] = _m
    
    return m
```
